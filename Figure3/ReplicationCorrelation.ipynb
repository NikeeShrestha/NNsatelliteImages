```python
import pandas as pd
import seaborn as sns
# from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
from scipy.stats import spearmanr, pearsonr
```

```python
df=pd.read_csv('../Data/HYBRID_HIPS_V3.5_ALLPLOTS.csv')
```

```python
df.columns
```

```python
df=df[df.location != 'North Platte3' ]
df=df[df.location != 'North Platte2' ]
df=df[df.location != 'North Platte1' ]
```

```python
df=df[['location', 'genotype', 'yieldPerAcre','nitrogenTreatment']]
df
# df=pd.DataFrame(df['genotype'].value_counts())
```

```python
df.set_index(['location', 'genotype', 'nitrogenTreatment'], inplace=True)
```

```python
my_dict = {'Rep1': [], 'Rep2': []}

# Iterate over DataFrame rows
for index, row in df.iterrows():
    rep1_indices = [item['Index'] for item in my_dict['Rep1']]
    rep2_indices = [item['Index'] for item in my_dict['Rep2']] 
    if index not in rep1_indices:
        my_dict['Rep1'].append({'Index': index, 'yield_Rep1': row['yieldPerAcre']})
    elif index not in rep2_indices:
        my_dict['Rep2'].append({'Index': index, 'yield_Rep2': row['yieldPerAcre']})
```

```python
rep1_df = pd.DataFrame(my_dict['Rep1'])
rep2_df = pd.DataFrame(my_dict['Rep2'])

# Merge the DataFrames on the index column
merged_df = pd.merge(rep1_df, rep2_df, on='Index', how='outer')
```

```python
merged_df
```

```python
for index, row in merged_df.iterrows():
    merged_df.at[index,'location']=row['Index'][0]
    merged_df.at[index,'genotype']=row['Index'][1]
    merged_df.at[index,'Nitrogen']=row['Index'][2]
    # print(row['Index'][0])
```

```python
sns.scatterplot(data=merged_df, x='yield_Rep1', y='yield_Rep2', hue='location', style='Nitrogen')
```

```python
merged_df['env']=merged_df['location']+'_'+merged_df['Nitrogen']
merged_df=merged_df.dropna()
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Nimbus Sans",
    "font.size":24,
    'axes.linewidth':2,
    'xtick.major.size':4,
    'ytick.major.size':4,
    'xtick.major.width':2,
    'ytick.major.width':2
    
})
fig = plt.figure(figsize=(30,45))
num=1
colors = {'Scottsbluff':'darkorange','Lincoln':'royalblue','Missouri Valley':'indigo','Ames':'plum', 'Crawfordsville':'olive'}
for i in merged_df.env.unique():
    tempdf=merged_df[merged_df['env']==i]
    myax=fig.add_subplot(5,3,num)
    sns.regplot(data=tempdf, x='yield_Rep1', y='yield_Rep2', ci=None, scatter_kws={'s':200, 'color':'red', 'alpha':1, 
                                                                                   'edgecolor':'black', 'linewidth': 2}
                ,line_kws={"color": "black"})
    
    location=str(i).split('_')[0]
    nitrogen=str(i).split('_')[1]
    plt.title(f'Location:{location}   Nitrogen:{nitrogen}')
    r, pval = pearsonr(tempdf.yield_Rep1.tolist(),tempdf.yield_Rep2.tolist())
    length=len(tempdf.yield_Rep1.tolist())
    myax.annotate(r"$R^2$={0:.2f}(n={1:.0f})".format(r*r, length), (min(tempdf['yield_Rep1']), max(tempdf['yield_Rep2'])-7), size=24)
    ax=plt.gca()
    xticks=ax.get_xticks()
    ax.set_xticks(xticks)
    ax.set_xticklabels([round(x) for x in xticks])
    
    yticks=ax.get_yticks()
    ax.set_yticks(yticks)
    ax.set_yticklabels([round(y) for y in yticks])
    
    plt.ylabel(r'Replication 2-Yield (lbs/acre)')
    plt.xlabel(r'Replication 1-Yield (lbs/acre)')
    
    num=num+1
    
# plt.savefig('SupplementaryRepCorrelation.png', dpi=350)
# plt.savefig('SupplementaryRepCorrelation.svg', dpi=350)

plt.show()
```
